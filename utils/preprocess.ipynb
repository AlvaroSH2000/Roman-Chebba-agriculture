```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('seaborn-v0_8-whitegrid')

path = "../data/Complete_dataset.xlsx"

df_other_raw = pd.read_excel(path, sheet_name="Other sites")
df_other = pd.DataFrame()
df_other["FID"] = df_other_raw["FID"]
```

# Primer mirem la correlació de Pearson i Spearman de tots els features

```python
df_other["Type"] = df_other_raw["Type"].map({"villa": 1, "farm": 0})
df_other["River_dist"] = df_other_raw["River_dist"]
df_other["Road_dist"] = df_other_raw["Road_dist"]
df_other["Near_500"] = df_other_raw["Near_500"]
df_other["Near_750"] = df_other_raw["Near_750"]
df_other["Near_1000"] = df_other_raw["Near_1000"]
df_other["Near_3000"] = df_other_raw["Near_3000"]
df_other["Landform_agri_score"] = df_other_raw["Landform_agri_score"]
df_other["Soil_agri_score"] = df_other_raw["Soil_agri_score"]
df_other["Aquifer_productivity"] = df_other_raw["Aquifer_productivity"]
```

```python
display(abs(df_other.corr('pearson')).style.background_gradient(cmap='coolwarm'))
display(abs(df_other.corr('spearman')).style.background_gradient(cmap='coolwarm'))
```

# Mirem especificament els d'hidrogeologia, i decidim quedar-nos amb el `Near_3000`

```python
display(abs(df_other[["Type", "River_dist", "Near_500", "Near_750", "Near_1000", "Near_3000"]].corr('pearson')).style.background_gradient(cmap='coolwarm'))
display(abs(df_other[["Type", "River_dist", "Near_500", "Near_750", "Near_1000", "Near_3000"]].corr('spearman')).style.background_gradient(cmap='coolwarm'))
df_other_clean = df_other.drop(columns=["River_dist", "Near_500", "Near_750", "Near_1000", "Landform_agri_score"])
```

# Guardem la base de dades neta

```python
df_other_clean.to_excel("../data/processed/other_sites_processed.xlsx", index=False)
df_other_clean.to_csv("../data/processed/other_sites_processed.csv", index=False)
```

# Guardem la base de dades neta pels sites de Chebba que sabem segur

```python
path = "../data/Complete_dataset.xlsx"

df_chebba_raw = pd.read_excel(path, sheet_name="Chebba sites")
df_chebba = pd.DataFrame()
df_chebba["ID"] = df_chebba_raw["ID"]

df_chebba["Type"] = df_chebba_raw["Type"]
df_chebba["Road_dist"] = df_chebba_raw["Road_dist"]
df_chebba["Near_3000"] = df_chebba_raw["Near_3000"]
df_chebba["Soil_agri_score"] = df_chebba_raw["Soil_agri_score"]
df_chebba["Aquifer_productivity"] = df_chebba_raw["Aquifer_productivity"]
df_chebba = df_chebba[df_chebba["Type"].notna()]

df_chebba.to_excel("../data/processed/combined_sites_processed.xlsx", index=False)
display(abs(df_chebba.drop("ID", axis=1).corr('pearson')).style.background_gradient(cmap='coolwarm'))
display(abs(df_chebba.drop("ID", axis=1).corr('spearman')).style.background_gradient(cmap='coolwarm'))
```

# Guardem la base de dades neta pels sites probables de Chebba (+ els que sabem segur)

```python
df_chebba_raw = pd.read_excel(path, sheet_name="Chebba sites")
df_chebba = pd.DataFrame()
df_chebba["ID"] = df_chebba_raw["ID"]

df_chebba["Type"] = df_chebba_raw["Type_probable"]
df_chebba["Road_dist"] = df_chebba_raw["Road_dist"]
df_chebba["Near_3000"] = df_chebba_raw["Near_3000"]
df_chebba["Soil_agri_score"] = df_chebba_raw["Soil_agri_score"]
df_chebba["Aquifer_productivity"] = df_chebba_raw["Aquifer_productivity"]
df_chebba = df_chebba[df_chebba["Type"].notna()]

df_chebba.to_excel("../data/processed/combined_probable_sites_processed.xlsx", index=False)
display(abs(df_chebba.drop("ID", axis=1).corr('pearson')).style.background_gradient(cmap='coolwarm'))
display(abs(df_chebba.drop("ID", axis=1).corr('spearman')).style.background_gradient(cmap='coolwarm'))
```

# Extra: Mirem la correlació quan ajuntem tots els sites de other_sites + Chebba probable

```python
df = pd.concat([df_other.drop(columns=["FID"]), df_chebba.drop(columns=["ID"])], ignore_index=True)

display(abs(df.corr('pearson')).style.background_gradient(cmap='coolwarm'))
display(abs(df.corr('spearman')).style.background_gradient(cmap='coolwarm'))
```
