```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('seaborn-v0_8-whitegrid')

path = "../data/Complete_dataset.xlsx"

df_other_raw = pd.read_excel(path, sheet_name="Other sites")
df_other = pd.DataFrame()
df_other["FID"] = df_other_raw["FID"]
```

```python
df_other["Type"] = df_other_raw["Type"].map({"villa": 1, "farm": 0})
df_other["River_dist"] = df_other_raw["River_dist"]
df_other["Road_dist"] = df_other_raw["Road_dist"]
df_other["Near_500"] = df_other_raw["Near_500"]
df_other["Near_750"] = df_other_raw["Near_750"]
df_other["Near_1000"] = df_other_raw["Near_1000"]
df_other["Near_3000"] = df_other_raw["Near_3000"]
df_other["Landform_agri_score"] = df_other_raw["Landform_agri_score"]
df_other["Soil_agri_score"] = df_other_raw["Soil_agri_score"]
df_other["Aquifer_productivity"] = df_other_raw["Aquifer_productivity"]
```

```python
display(abs(df_other.corr('pearson')).style.background_gradient(cmap='coolwarm'))
display(abs(df_other.corr('spearman')).style.background_gradient(cmap='coolwarm'))
```

```python
df_other = df_other.drop(columns=["River_dist", "Near_500", "Near_750", "Near_1000"])
df_other.to_excel("../data/processed/other_sites_processed.xlsx", index=False)
```
